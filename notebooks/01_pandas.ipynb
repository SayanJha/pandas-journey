```python
data = {
    "name": ["John", "Anna", "Peter", "linda"],
    "Age": [28, 34, 29, 42],
    "City": ["New Work", "Paris", "Berlin", "London"],
    "Salary": [65000, 70000, 62000, 85000],
}
```

```python
df = pd.DataFrame(data)
print(type(df))  # Pandas dataframe
```

```python
col_series = df["Age"]
col_df = df[["Age"]]

print(type(col_series))  # Series
print(type(col_df))  # dataframe
print("Shapes")

print(col_series.shape)  # (4, )
print(col_df.shape)  # (4, 1)

weights = np.array([[2]])  # (1, 1)
print(weights.shape)

print("\n")
# * print(col_series @ weights)
#! Exception: Dot product shape mismatch, (4,) vs (1, 1)

print(col_df @ weights)
# :)
```
